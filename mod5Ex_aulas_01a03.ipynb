````python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(f'Quantidade de linhas duplicadas: {sinasc.duplicated().sum()}')

# Como não há duplicados, o número de registros não duplicados vai ser = o número de registros
print(f'Quantidade de registros: {sinasc.shape[0]}')


sinasc
````

````python
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# 3) seu código aqui
colum_interesse = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
    'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]

colum_interesse.isna().sum()
````

````python
# 4) seu código aqui
colum_interesse = colum_interesse.dropna(subset=['APGAR5'])
print(f'Número de linhas: {colum_interesse.shape[0]}')
colum_interesse.isna().sum()
````

````python
# 5) seu código aqui
colum_interesse['ESTCIVMAE'].replace(9, 'não preenchido', inplace=True)
colum_interesse['CONSULTAS'].replace(9, 'não preenchido', inplace=True)
````

````python
# 6) Seu código aqui
colum_interesse['QTDFILVIVO'].fillna(0, inplace=True)
````

````python
# 7) seu código aqui. Justificativa: Numa variável como IDADEMAE não faz sentido eu repetir a idade anterior
#já que ela prtence a outra pessoa, então opitei por não preenchido
colum_interesse.fillna('não preenchido', inplace=True)
````

````python
# 8) seu código aqui
colum_interesse.loc[colum_interesse['APGAR5'] >= 8, 'Classificação'] = 'normal'
colum_interesse.loc[(colum_interesse['APGAR5'] >= 6) & (colum_interesse['APGAR5'] <= 7),
           'Classificação'] = 'asfixia leve'
colum_interesse.loc[(colum_interesse['APGAR5'] >= 4) & (colum_interesse['APGAR5'] <= 5),
           'Classificação'] = 'asfixia moderada'
colum_interesse.loc[(colum_interesse['APGAR5'] <= 3), 'Classificação'] = 'asfixia severa'

print(colum_interesse['Classificação'].value_counts())
colum_interesse.head(21)
````
