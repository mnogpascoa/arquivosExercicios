````python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax=axes[0, 0], x='data_ref', hue='sexo', data=df_analise)
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[0, 1], x='data_ref', hue='posse_de_veiculo', data=df_analise)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[1, 0], x='data_ref', hue='posse_de_imovel', data=df_analise)
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[1, 1], x='data_ref', hue='tipo_renda', data=df_analise)
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')
````

````python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax=axes[0, 0], x='data_ref', hue='educacao', data=df_analise)
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[0, 1], x='data_ref', hue='estado_civil', data=df_analise)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[1, 0], x='data_ref', hue='tipo_residencia', data=df_analise)
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=45, ha='right')


sns.countplot(ax=axes[1, 1], x='data_ref', hue='mau', data=df_analise)
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')
````
