Dica: você pode usar o método ```.str``` para transformar o tipo da variável em string, em seguida um método como [**slice()**](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.slice.html) ou [**split()**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.split.html).
4. Indique a frequência de cada ano do campo construído no item (3).

````python
# 1) Seu código aqui
url = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Estado' 
r = requests.get(url)
print(r.status_code)

data_json = r.json()
RJ = pd.DataFrame(data_json['items'])
RJ
````

````python
# 2) Seu código aqui
RJ['status'].value_counts()
````

````python
# 3) Seu código aqui
lista_datas = RJ['data_status'].astype(str).tolist()
lista_anos = [data.split('/')[2] for data in lista_datas]
df_anos = pd.DataFrame(lista_anos)

RJ = pd.concat([RJ, df_anos], axis=1)
RJ.columns.values[-1] = 'Ano'
RJ
````

````python
RJ['Ano'].value_counts()
````

Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?

````python
#1) Seu código aqui
print('Escolha a UF e o tipo interessado:')
uf = input()
es_mu = input()
````

````python
#1) Seu código aqui
url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={es_mu}' 
r = requests.get(url)
print(r.status_code)

data_json = r.json()
df = pd.DataFrame(data_json['items'])

df
````

````python
# 2) Seu código aqui
contagem_arquivado = df[df['status'] == 'Arquivado por decurso de prazo'].shape[0]
print(f'Número de solicitações Arquivado por decurso de prazo: {contagem_arquivado}')
````

````python
# 3) Seu código aqui
url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=BA&tipo_interessado=Município' 
r = requests.get(url)
print(r.status_code)

data_json = r.json()
df = pd.DataFrame(data_json['items'])

df
````

````python
contagem_status = df['interessado'].value_counts()
valor_mais_frequente = contagem_status.idxmax()
quantidade_repeticoes = contagem_status.max()
print(f'Mais frequente e quantidade de vezes que se repete respectivamente: {valor_mais_frequente}, {quantidade_repeticoes}')
````

````python
# 4) Seu código aqui
url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=BA&tipo_interessado=Estado' 
r = requests.get(url)

data_json = r.json()
df = pd.DataFrame(data_json['items'])

df.to_csv('./dataframe_bahia.csv')
````
