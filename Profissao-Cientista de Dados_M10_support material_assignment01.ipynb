2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
scatter = sns.scatterplot(x='tempo_emprego', y='renda', hue='educacao', data=df)
plt.xlabel('Tempo de Emprego (anos)')
plt.ylabel('Renda')

plt.title('Análise da renda por tempo de emprego e educação')

scatter.legend(title='Educação')
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
df['renda_log'] = np.log(df['renda'])
scatter = sns.scatterplot(x='tempo_emprego', y='renda_log', hue='educacao', data=df)
plt.xlabel('Tempo de Emprego (anos)')
plt.ylabel('Log_Renda')

plt.title('Análise da renda por tempo de emprego e educação')

scatter.legend(title='Educação')
````
